```python
#filtrando os dados especificando uma condição
dados.loc[dados['credit_amount'] > 18000]
```

```python
#atribuindo o resultado anterior a uma variável
credito = dados.loc[dados['credit_amount'] > 18000]
print(credito)
```

```python
#renomeando células
dados.rename(columns={"duration": "duração", "purpose":"propósito"})
#para a alteração ser permanente, é necessário escrever o código da seguinte maneira: dados.rename(columns={"duration":"duração","purpose":"propósito"},inplace=True)
```

```python
#excluindo coluna
dados.drop('checking_status',axis=1,inplace=True)
print(dados)
```

```python
#verificando dados nulos
dados.isnull()
```
